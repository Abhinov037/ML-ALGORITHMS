```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns  # Ensure seaborn is imported
```

```python
x=np.arange(10)
y=(x-5)**2
print(x,y)
```

```python
# Given a Function f(x), we want to find the value of x that minimizes f
```

```python
# Use seaborn to plot the data
sns.set(style="darkgrid")
plt.plot(x, y)
plt.ylabel("f(x)")
plt.xlabel("x")
plt.title("Plot of f(x) vs x")
plt.show()
```

```python
# Gradient descent example
error = []

# Initial guess
x = 0

# Learning rate
lr = 0.1

# Number of iterations
iterations = 50

# Gradient descent loop
for i in range(iterations):
    grad = 2 * (x - 5)  # Gradient of the function f(x) = (x-5)^2
    x = x - lr * grad
    error.append((x - 5) ** 2)

# Plotting the error over iterations
plt.scatter(range(iterations), error)
plt.xlabel("Iteration")
plt.ylabel("Error")
plt.title("Error over iterations in gradient descent")
plt.show()
```
